```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
import matplotlib.pylab as pylab

# Set default image size and font sizes
pylab.rcParams['figure.figsize'] = 12, 7
pylab.rcParams['font.size'] = 15
pylab.rcParams['xtick.labelsize'] = 15
pylab.rcParams['ytick.labelsize'] = 15
pylab.rcParams['axes.labelsize'] = 15

#Model	Site	Climate	National	Dirindex	Dirint	Disc	Yang5	Yang4	Engerer2	Engerer3	Engerer4	Nested	Erbs	ED	Ohm	Louche	Boland


# Load the data
obs_file = 'ALLSKY.csv'
columns = ['Model','Proposed','Yang5','Yang4','Engerer2','Engerer3', 'Engerer4','Nested','Erbs', 'ED', 'OH','Louche','Boland']
obs = pd.read_csv(obs_file, header=None, skiprows=1, names=columns)

# Plot the bar graph
ax = obs.plot.bar(figsize=(16, 6), width=0.8)
plt.xlabel('Parameter', fontsize=15)
plt.ylabel('Mean rank', fontsize=15)
ax.set_ylim([1, None])
plt.xticks([0, 1, 2], ['Kd', 'DIF', 'DNI'], rotation=0)
ax.legend(title='Model', title_fontsize=15, loc='center left', bbox_to_anchor=(1, 0.5))

# Show the plot
plt.show()
```
